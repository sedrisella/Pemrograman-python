````python
import matplotlib.pyplot as plt

# Data
bands = ['Slipknot', 'Linkin Park', 'Avenged Sevenfold', 'Bring Me the Horizon', 'Green Day', 'Muse', 'As I Lay Dying', 'Trivium', 'Limp Bizkit', 'Skid Row', 'Bon Jovi']
fans = [6603189, 17418362, 5410274, 4428374, 15692462, 8528417, 996422, 1880758, 5910689, 2492044, 15581843]

# Create a bar chart
plt.bar(bands, fans)

# Add labels
plt.title('Jumlah pendengar perbukan di spotify')
plt.xlabel('Nama Band')
plt.ylabel('Pendengar per bulan')

# Show the plot
plt.show()
````
