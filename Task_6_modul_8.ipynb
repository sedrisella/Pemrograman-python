#Machine Learning : Prediksi Harga Rumah
#dengan Regresi Linear (Multi Variabel)

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df=pd.read_csv('datarumah.csv')
```

```python
#at digunakan untuk mengisi value(0)index ke 2,kolom kamar tidur
#cell yang kosong diisi dengan nilai 0
df.at[2,'kamartidur']=0
df
```

```python
#mencari nilai median dari coloum kamar tidur
df.kamartidur.median()
```

```python
 #membulatkan nilai ke bawah
import math
nilai_median_kamartidur= math.floor(df.kamartidur.median())
nilai_median_kamartidur
```

```python
df.kamartidur.fillna(nilai_median_kamartidur)
```

```python
df.kamartidur=df.kamartidur.fillna(nilai_median_kamartidur)
df
```

```python
reg =linear_model.LinearRegression()
reg.fit(df[['luas','kamartidur','kamarmandi']],df.harga)
```
